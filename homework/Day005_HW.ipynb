```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

'''
Your Code
'''
import requests
import json
import pandas as pd 
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text
response_pop=pd.read_json(response)
print(len(response_pop),response_pop.columns)
```
